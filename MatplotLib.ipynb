# Simple Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 4, 6]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

**Unit 2: Basics of Plotting**

```python
# A simple plot is the most basic visualization in Matplotlib, usually a line plot showing a relationship between two sets of data (X and Y).
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.show()
```

# plt.plot() Basics

* plt.plot() is used to draw line plots, the most basic and commonly used plot in Matplotlib.

* Accepts lists, tuples, NumPy arrays

* Automatically connects data points with a line

* Can customize:

- Line style (--, -., :)

- Marker (o, *, s)

- Color (r, g, b)

```python
# plt.plot(x, y, format_string)
plt.plot([1,2,3], [1,4,9], 'ro--')   # red, circle marker, dashed line
plt.show()
"""
Explanation

'r' = red

'o' = circle marker

'--' = dashed line """
```

# Labels, Titles, Legends
Text elements added to graphs to improve readability.


* plt.title()  Name of the graph

* plt.xlabel()  Text for X-axis

* plt.ylabel()  Text for Y-axis

* plt.legend()  Identifies multiple plotted lines


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 18, 22]

plt.plot(x, y1,label="Sales 2023")
plt.plot(x, y2, label="Sales 2024")

plt.title("Sales Comparison")
plt.xlabel("Quarter")
plt.ylabel("Revenue (in Lakhs)")
plt.legend()
plt.show()
```

**Unit 3: Plot Types**

Line Plot

* Definition:

A line plot displays data points connected by straight lines.
It is used to show trends over time or continuous data.

* Theory

Best for time series, trend analysis, changes over intervals

Uses plt.plot()

* Syntax
plt.plot(x, y)

```python
import matplotlib.pyplot as plt

months = [1, 2, 3, 4, 5]
sales = [20, 25, 30, 28, 35]

plt.plot(months, sales)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

Bar Chart
* Definition

A bar chart represents categorical data with rectangular bars.

* Theory

Good for comparison between categories

Bars can be vertical (plt.bar()) or horizontal (plt.barh())

* Syntax
plt.bar(categories, values)

```python
subjects = ["Math", "Science", "English"]
marks = [85, 90, 78]

plt.bar(subjects, marks)
plt.title("Student Marks")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.show()
```

Histogram
* Definition

A histogram shows the distribution of a dataset by dividing values into bins (ranges).

* Theory

Used to visualize frequency distribution

Helps find:

Spread

Skewness

Outliers

Normal distribution

* ax
plt.hist(data, bins=number)

```python
import numpy as np

data = np.random.randn(200)

plt.hist(data, bins=20,color="red",)
plt.title("Data Distribution")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.show()
```

Scatter Plot
* Definition

A scatter plot shows the relationship between two variables using dots.

* Theory

Useful for:

Correlation

Patterns

Trends

Clustering

Plots individual (x, y) points

* Syntax
plt.scatter(x, y)

```python
height = [150, 155, 160, 165, 170]
weight = [50, 55, 60, 65, 70]

plt.scatter(height, weight)
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

Box Plot
* Definition

A box plot visualizes data spread using:

Minimum

Q1 (25%)

Median (50%)

Q3 (75%)

Maximum

Outliers

```python
data = [25, 30, 28, 35, 40, 38, 50, 55, 60]

plt.boxplot(data)
plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

Pie Chart
* Definition

A pie chart shows proportions of categories as slices of a circle.

* Theory

Best for percentage distribution

Not ideal for too many categories

* Syntax
plt.pie(values, labels=labels, autopct="%1.1f%%")

```python
labels = ["Chrome", "Firefox", "Edge", "Safari"]
market_share = [60, 15, 10, 15]

plt.pie(market_share, labels=labels, autopct="%1.1f%%")
plt.title("Browser Market Share")
plt.show()
```

**Unit 4: Customization**


1.colors

Matplotlib supports many color formats:

Single-letter colors: 'r', 'g', 'b', 'k'

Color names: "red", "green", "blue"

Hex colors: "#FF5733"

RGB values: (0.4, 0.2, 0.6)

Also supports built-in styles like:

"ggplot"

"seaborn"

"classic"

"bmh"

```python
plt.plot(x, y, color="red")
plt.style.use("ggplot")
```

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")

x = [1, 2, 3, 4]
y = [10, 15, 13, 20]

plt.plot(x, y, color="purple")
plt.title("Customized Color & Style")
plt.show()
```

2.Markers and Line Styles

plt.plot(x, y, marker='o', linestyle='--', color='orange')


```python
x = [1, 2, 3, 4]
y = [5, 7, 6, 9]

plt.plot(x, y, marker='*', linestyle='--', color='green')
plt.title("Markers & Line Styles Example")
plt.show()
```

3.Grid Customization

plt.grid(True)

plt.grid(color='gray', linestyle='--', linewidth=0.5)



```python
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.grid(True, linestyle='--', color='gray')
plt.title("Customized Grid Example")
plt.show()
```

Axes limits define the visible range of X and Y axes.
Ticks are the values displayed on axes.

Theory

Use plt.xlim() and plt.ylim() to control axis boundaries

Use plt.xticks() and plt.yticks() to format ticks

```python
x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.plot(x, y)

plt.xlim(0, 6)
plt.ylim(0, 30)

plt.xticks([0, 2, 4, 6])
plt.yticks([0, 10, 20, 30])

plt.title("Axes Limits & Ticks Customization")
plt.show()
```
