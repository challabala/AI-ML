Why do we need TF‚ÄìIDF?

In raw text:

Words are just strings ‚Üí ML models cannot work with them.

TF‚ÄìIDF converts text into numbers that reflect:

How frequently a word appears in a document

How unique that word is across all documents

This helps models differentiate important terms from unimportant ones.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample documents
docs = [
    "Machine learning is amazing",
    "Deep learning and machine learning",
    "Machine learning applications are growing"
]

# Create TF-IDF vectorizer
tfidf = TfidfVectorizer()

# Fit and transform the documents
tfidf_matrix = tfidf.fit_transform(docs)

# Show feature names
print("Features:", tfidf.get_feature_names_out())

# Show TF-IDF matrix
print(tfidf_matrix.toarray())
```

```python
# Example - 2
from sklearn.feature_extraction.text import TfidfVectorizer

docs = [
    "natural language processing and machine learning",
    "deep learning improves natural language tasks",
    "language models are important for processing text"
]

tfidf = TfidfVectorizer()
matrix = tfidf.fit_transform(docs)

print(tfidf.get_feature_names_out())
print(matrix.toarray())
```
