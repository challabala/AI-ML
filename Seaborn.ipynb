<h1>What is Seaborn</h1>
<p>Seaborn is a Python data visualization library built on top of Matplotlib, designed to create beautiful, statistical, and informative plots with very little code.</p>

<ul> <li>Seaborn simplifies complex visualizations.</li> <li>It comes with attractive default themes and color palettes.</li> <li>It integrates tightly with Pandas DataFrames, meaning it can directly plot data from columns.</li> <li>Seaborn includes built-in datasets (like <b>tips</b>, <b>iris</b>, <b>flights</b>) for practicing data visualization.</li> </ul>


<ul> <li>Easy plotting of statistical graphs (histogram, KDE, boxplot, violin, pairplot, heatmap).</li> <li>Works naturally with Pandas.</li> <li>Requires less code compared to Matplotlib.</li> <li>Advanced patterns for visualizing distributions.</li> </ul>

**Unit 3: Relational Plots**



```python
# 1. scatterplot()
# scatterplot() creates a scatter plot, showing the relationship between two numerical
# sns.scatterplot(x, y, data=None, hue=None, style=None, size=None)


import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.scatterplot(x="sepal_length", y="petal_length", data=iris)
plt.show()
```

```python
sns.scatterplot(
    x="sepal_length",
    y="petal_length",
    data=iris,
    hue="species"
)
plt.show()
```

```python
# lineplot()
# lineplot() is used to visualize trends over time or relationship between continuous
# sns.lineplot(x, y, data=None, hue=None, style=None)

import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

sns.lineplot(x="year", y="passengers", data=flights)
plt.show()
```

```python
sns.lineplot(
    x="year",
    y="passengers",
    data=flights,
    hue="month"
)
plt.show()
```

```python
# **3. relplot() (Facet Interface)
"""sns.relplot(
    x, y,
    data=None,
    hue=None,
    col=None,
    row=None,
    kind="scatter"
)"""
import seaborn as sns

iris = sns.load_dataset("iris")

sns.relplot(
    x="sepal_length",
    y="petal_length",
    hue="species",
    col="species",
    kind="scatter",
    data=iris
)
```

```python
flights = sns.load_dataset("flights")

sns.relplot(
    x="year",
    y="passengers",
    hue="month",
    kind="line",
    col="month",
    col_wrap=4,
    data=flights
)
```

**Unit 4: Categorical Plots (Seaborn)**

```python
# 1. barplot()
# sns.barplot(x, y, data=None, hue=None)
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
# 2. countplot()
# countplot() displays the count of occurrences of each category.
# sns.countplot(x, data=None, hue=None)
sns.countplot(x="day", data=tips)
plt.show()
```

```python
# 3. boxplot()
# boxplot() visualizes the distribution of numerical data, highlighting:
# sns.boxplot(x, y, data=None, hue=None)

sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
# 4. violinplot()
# violinplot() shows the distribution of data using:
sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
# 5. stripplot()
# stripplot() draws a scatter of all data points for each category.

sns.stripplot(x="day", y="total_bill", data=tips)
plt.show()
```

***Unit 5: Distribution Plots (Seaborn)***

```python
# 1. histplot()
# histplot() creates a histogram, showing how many values fall into different ranges (bins).
# sns.histplot(data=None, x=None, bins=None, kde=False, hue=None)
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.histplot(x="total_bill", data=tips)
plt.show()
```

```python
# 2. distplot() (LEGACY)
sns.distplot(tips["total_bill"], kde=True)
plt.show()
```

```python
# 3. kdeplot()
# kdeplot() shows a Kernel Density Estimate, which is a smooth curve representing data distribution.
# sns.kdeplot(data=None, x=None, hue=None, fill=False)
sns.kdeplot(x="total_bill", data=tips, fill=True)
plt.show()
```

```python
# 4. jointplot()
# jointplot() creates a combined plot of:
# Relationship between two variables (scatter/hist/kde)
# Along with individual distributions on top and right

# sns.jointplot(x=None, y=None, data=None, kind="scatter")


iris = sns.load_dataset("iris")

sns.jointplot(x="sepal_length", y="petal_length", data=iris, kind="scatter")
plt.show()
```

```python
# 5. pairplot()
# pairplot() creates multiple plots for every pair of numerical columns in the dataset.
# sns.pairplot(data=None, hue=None)

sns.pairplot(iris, hue="species")
plt.show()
```

** Unit 6: Matrix & Heatmap Plots ** *italicized text*



```python
# 1. heatmap()
# heatmap() displays data in a matrix format, where each cell is represented by a color intensity based on its value.
# sns.heatmap(data, annot=False, cmap=None, linewidths=0.5)

import seaborn as sns
import matplotlib.pyplot as plt

# Corrected pivot method using keyword arguments
pivot_data = flights.pivot(index="month", columns="year", values="passengers")

sns.heatmap(pivot_data, annot=True, cmap="YlGnBu")
plt.show()
```

```python
# 2. clustermap()
# clustermap() creates a clustered heatmap with hierarchical clustering, adding dendrograms (tree structures) on the sides.
# sns.clustermap(data, cmap=None, standard_scale=None)
sns.clustermap(pivot_data, cmap="coolwarm")
plt.show()
```

**Unit 7: Styling & Themes**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
tips = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.set_style("dark")
sns.histplot(tips["total_bill"])
plt.show()
```

```python
sns.set_palette("husl")
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
sns.set_context("talk")
sns.set_palette("coolwarm")

tips = sns.load_dataset("tips")

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Total Bill vs Tip")
plt.show()
```
