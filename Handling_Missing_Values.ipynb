```python
# Another data set
data = pd.read_csv('/content/null_values_dataset.csv')
data.head()
```

```python
data[["Age","Salary"]].isnull()
```

```python
data.isnull().sum()
```

```python
data.fillna(0)
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Salary'] = data['Salary'].fillna(data['Salary'].mean())
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data['City'] = data['City'].fillna(data['City'].mode()[0])
```

```python
data.isnull().sum()
```

```python
data.dropna(axis=1,inplace=True)
```
